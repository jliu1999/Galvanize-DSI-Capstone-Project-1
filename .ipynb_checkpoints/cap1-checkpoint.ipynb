## 12, Checking normality with the Shapiro-Wilk Normality Test

<br>

**One example using Anderson-Darling Test:**

```python
# An example of Anderson-Darling Test, will not use it for analysis
data = df['LifeExpectancy']
result = stats.anderson(data)
print('stat: {0:.3f}'.format(result.statistic))
print(result.critical_values)
print(result.significance_level)
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
        print('Probably Gaussian at the %.1f%% level' % (sl))
    else:
        print('Probably not Gaussian at the %.1f%% level' % (sl))
```

***Anderson-Darling Test proves that Life Expectancy is not normally distributed.***

<br>

### Performing the Shapiro-Wilk Test for normality


    H0: the sample is normally distributed.
    H1: the sample is not normally distributed.


```python
# Just a try
data = df['LifeExpectancy']
stat, p = stats.shapiro(data)
print('stat={0:.3f}, p={1:.3f}' .format(stat, p))
if p > 0.05:
    print('Probably normally distributed')
else:
    print('Probably not normally distributed')
```

***Shapiro-Wilk Test also proves that Life Expectancy is not normally distributed.***

<br>

**Writing a function to test the normality for all features**

```python
def test_normality(data):
    for col in data.columns[1:]:
        stat, p = stats.shapiro(data[col].dropna())
        print('For {0}, stat={1:.3f}, p={2:.3f}'.format(col, stat, p))
        if p > 0.05:
            print(f'{col} is probably normally distributed.\n')
        else:
            print(f'{col} is probably not normally distributed.\n')
```

```python
test_normality(df)
```

### Actually doing the Spearman Correlation Test

The Spearman's rank correlation coefficient is often used as a test statistic in a statistical hypothesis test to establish whether two variables may be regarded as statistically dependent. Unlike the Pearson's Correlation test which assumes the random variables are normally distributed, the Spearman's test does not rely on any assumptions on the distributions of the variables, namely this test is non-parametric. 

This test is based on the following assumptions:

**1) Observations in each sample are independent and distributed identically.**

**2) Observations in each sample are ranked.**

**The hypotheses are:**

    H0: the two samples do not have any correlation.
    H1: the two samples are correlated.

**Don't know if the samples are positively correlated or negativelly correlated, so it is a two-sided test.**

<br>

**Start with some trials, using the Spearman's Rank Correlation Test to test if there are correlations among LifeExpectancy and others.**

```python
data1 = df['LifeExpectancy']
data2 = df['uhcCoverage']

stat, p = stats.spearmanr(data1, data2)
print('stat={0:.3f}, p={1:.3f}'.format(stat, p))

# interpret the significance
alpha = 0.05
if p > alpha:
    print('Probably independent')
else:
    print('Probably dependent')
```

***The statistical test reports a strong positive correlation with a value of 0.88. The p-value is close to zero, which means that the likelihood of observing the data given that the samples are uncorrelated is very unlikely at 95% confidence, and that we can reject the null hypothesis that the samples are uncorrelated.***

```python
data = df[df['HHExpdt10'].notnull()][['LifeExpectancy', 'HHExpdt10']]

stat, p = stats.spearmanr(data['LifeExpectancy'], data['HHExpdt10'])
print('stat={0:.3f}, p={1:.3f}'.format(stat, p))
alpha = 0.05
if p > alpha:
    print('Probably independent')
else:
    print('Probably dependent')
```

***Although the value of 0.302 shows some weak correlation, the p-value is bigger than 0.05. We can not reject the null hypothesis. Life Expectancy is not related with the Household Healthcare Expenditure which is more than 10% percent of the household income.***

## 14, Calculating the 95% confidence interval

**Spearman Test is testing on non-parametric distributions, which means we don't have parameters for the distribution of the variable, so we also don't have pdf() or ppf() to calculate the confidence interval.** 

I found from the following resource the formula to calculate the 95% confidence interval:
<br>

### <B><center>$\tanh(\operatorname{arctanh}r\pm1.96/\sqrt{n-3})$</center></B>
<br>

Reference: https://stats.stackexchange.com/questions/18887/how-to-calculate-a-confidence-interval-for-spearmans-rank-correlation

**First, do some experiment with the first element from the correlation matrix:**

```python
r = 0.883
num = 183
stderr = 1.0 / math.sqrt(num - 3)
delta = 1.96 * stderr
lower = math.tanh(math.atanh(r) - delta)
upper = math.tanh(math.atanh(r) + delta)
print(f"The 95% confidence interval is ({lower:.3f}, {upper:.3f})")
```

***The result looks reasonable.***
<br>


### Writing a function to calculate the confidence intervals for all pairs of features

```python
# make a list of the labels of all relevent columns
col = df.columns[1:]
col
```

```python
def spearman_ci(data):
    dict_ = {}
    for i in range(len(data)):
        for j in range(i+1,len(data)):
            num = len(df[[col[i],col[j]]].dropna())
            stderr = 1.0 / math.sqrt(num - 3)
            delta = 1.96 * stderr
            r = data[i][j]
            lower = round(math.tanh(math.atanh(r) - delta), 3)
            upper = round(math.tanh(math.atanh(r) + delta), 3)
            dict_[f'{col[i]} / {col[j]}'] = (lower,upper)
    return dict_
```

## 16, Calculating Confidance Interval using Bootstraping


```python
df_le_uhc = df[['LifeExpectancy','uhcCoverage']]
df_le_uhc
```

```python
corr_test, p_test = stats.spearmanr(df_le_uhc.iloc[:, 0], df_le_uhc.iloc[:, 1])
corr_test, p_test
```

```python
np.random.seed(1)
bootstrap = []
for i in range(10000):
    # Have to bootstrap pairwise, so randomly sample the indices, that select the data from the dataframe
    boot = df_le_uhc.iloc[np.random.choice(df_le_uhc.shape[0], size=df_le_uhc.shape[0], replace=True)]
    corr_le_uhc, p_value = stats.spearmanr(boot)
    bootstrap.append(corr_le_uhc)
bootstrap
```

```python
left = np.percentile(bootstrap, 2.5)
right = np.percentile(bootstrap, 97.5)

print("The correlation is: {:.3f}".format(stats.spearmanr(df_le_uhc.iloc[:, 0], df_le_uhc.iloc[:, 1])[0]))
print(f"Bootstrap Confidence Interval for Correlation: {left:.3f}, {right:.3f}")
```

```python
fig, ax = plt.subplots(1, figsize=(12, 5))
ax.hist(bootstrap, bins=50, density=True, alpha=0.5)
ax.axvline(stats.spearmanr(df_le_uhc.iloc[:, 0], df_le_uhc.iloc[:, 1])[0], color='r')
ax.axvline(left, color='g')
ax.axvline(right, color='g')
ax.set_title("Confidence Interval by Bootstrapping", fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=15)
#ax.set_xlabel('Country', fontsize=15)
#ax.set_ylabel('Nurse Count Per 10,000 Population', fontsize=15)
plt.tight_layout()
plt.savefig('images/bootstrap.png');
```
