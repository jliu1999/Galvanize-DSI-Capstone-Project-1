```python
le_2015_copy = le_2015.copy()
```

Now we will use the Spearman's Rank Correlation Test to test if there is correlation between LifeExpectancy and others.

```python
# Example of the Spearman's Rank Correlation Test
data1 = df['LifeExpectancy']
data2 = df['uhcCoverage']

stat, p = stats.spearmanr(data1, data2)
print('stat={0:.4f}, p={1:.4f}'.format(stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```
